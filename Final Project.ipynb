```python
# for dataframe manipulations
import pandas as pd
# for Data Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Load Data**

```python
df = pd.read_csv('/content/insurance.csv')
df
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['sex'].replace(['female', 'male'], [0,1], inplace=True)
df['smoker'].replace(['yes', 'no'], [0,1], inplace=True)
df['region'].replace(['southwest','southeast','northwest','northeast'], [0,1,2,3], inplace=True)
df
```

```python
df['charges'] = df['charges'].astype(int)
df['bmi'] = df['bmi'].astype(int)
df
```

# **Model**

```python
X =  df.drop(["region","charges"],axis=1)
X
```

```python
y = df['charges']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

**Linear Regression**

```python
#Linear Regression
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()  


linear_regressor.fit(X_train, y_train)
```

```python
# prints y-intercept
print(linear_regressor.intercept_)

# prints the coefficient
print(linear_regressor.coef_)
```

```python
y_pred_lr = linear_regressor.predict(X_test)  
```

```python
from sklearn.metrics import r2_score
r2_lr = r2_score(y_test.values, y_pred_lr)
print(r2_lr)
```

```python
import numpy as np
plt.figure(figsize=(10,8))
plt.plot(np.arange(len(y_test)), y_test,label='Aktual')
plt.plot(np.arange(len(y_test)), y_pred_lr,label='Prediksi')
plt.title('Model Linear Regression')
plt.legend()
plt.show()
```

**Decision tree**

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()

dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)  
```

```python
r2_dt = r2_score(y_test.values, y_pred_dt)
print(r2_dt)
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.arange(len(y_test)), y_test,label='Aktual')
plt.plot(np.arange(len(y_test)), y_pred_dt,label='Prediksi')
plt.title('Model Decision Tree')
plt.legend()
plt.show()
```

**Random Forest**

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=19)

rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test) 

r2_rf = r2_score(y_test.values, y_pred_rf)
print(r2_rf)
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.arange(len(y_test)), y_test,label='Aktual')
plt.plot(np.arange(len(y_test)), y_pred_rf,label='Prediksi')
plt.title('Model Random Forest')
plt.legend()
plt.show()
```
